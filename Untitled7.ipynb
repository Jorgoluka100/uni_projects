```python
# Linking with Spark
import pyspark
from pyspark.sql import SparkSession

spark = SparkSession.builder \
                    .appName("Tutorial_DF") \
                    .master("local[*]") \
                    .config("spark.executor.memory", "4g") \
                    .config("spark.driver.memory", "2g") \
                    .config("spark.executor.cores", "2") \
                    .config("spark.sql.inMemoryColumnarStorage.compressed", "true") \
                    .getOrCreate()

spark
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = spark.read.format('csv').load('/content/drive/MyDrive/Files/games.csv', inferSchema = True, header = True)

# show table
df.show(truncate = True)

# show schema
df.printSchema()

# some info
print(df.count())
print(len(df.columns))
```

Data Pre-processing

```python
# Converting data to numerics in order to readable to models
# includes both categorical and string data
from pyspark.ml.feature import StringIndexer, OneHotEncoder
```

```python
df_country_code1 = StringIndexer(inputCol='country_code',outputCol='country_code_numeric').fit(df)
df = df_country_code1.transform(df)
encoder_country_code = OneHotEncoder(inputCol='country_code_numeric',outputCol="country_code_vector").fit(df)
df = encoder_country_code.transform(df)
df = df.drop('country_code', 'country_code_numeric')
```
