```python
from google.oauth2 import service_account
from googleapiclient.discovery import build
import re
```

```python
# Define the scopes and credentials file
SCOPES = ['https://www.googleapis.com/auth/documents.readonly']
SERVICE_ACCOUNT_FILE = 'credentials.json'
```

```python
def get_document_content(doc_id):
    """Fetch content from Google Docs using the API."""
    creds = service_account.Credentials.from_service_account_file(
        SERVICE_ACCOUNT_FILE, scopes=SCOPES)

    service = build('docs', 'v1', credentials=creds)
    document = service.documents().get(documentId=doc_id).execute()

    content = []
    for element in document.get('body').get('content'):
        paragraph = element.get('paragraph')
        if paragraph:
            for el in paragraph.get('elements'):
                text_run = el.get('textRun')
                if text_run:
                    content.append(text_run.get('content'))

    return ''.join(content)
```

```python
def parse_content(content):
    """Parse the content into a grid of characters based on coordinates."""
    grid = {}
    lines = content.splitlines()

    for line in lines:
        # Example line: "A 1 2"
        match = re.match(r'(\S) (\d+) (\d+)', line)
        if match:
            char = match.group(1)
            x = int(match.group(2))
            y = int(match.group(3))
            if y not in grid:
                grid[y] = {}
            grid[y][x] = char
```

```python
def parse_content(content):
    """Parse the content into a grid of characters based on coordinates."""
    grid = {}
    lines = content.splitlines()

    for line in lines:
        # Example line: "A 1 2"
        match = re.match(r'(\S) (\d+) (\d+)', line)
        if match:
            char = match.group(1)
            x = int(match.group(2))
            y = int(match.group(3))
            if y not in grid:
                grid[y] = {}
            grid[y][x] = char

    # Determine grid dimensions
    max_x = max(max(row.keys(), default=0) for row in grid.values())
    max_y = max(grid.keys(), default=0)

    # Create the grid with spaces
    result_grid = [[' ' for _ in range(max_x + 1)] for _ in range(max_y + 1)]

    # Fill the grid with characters
    for y, row in grid.items():
        for x, char in row.items():
            result_grid[y][x] = char

    return result_grid
```

```python
def print_grid(grid):
    """Print the grid."""
    for row in grid:
        print(''.join(row))
```

```python
def main(url):
    doc_id = url.split('/d/')[1].split('/')[0]
    content = get_document_content(doc_id)
    grid = parse_content(content)
    print_grid(grid)
```

```python
import gspread
from oauth2client.service_account import ServiceAccountCredentials
```

```python
def get_google_sheet_data(url):
    # Set up the credentials for Google Sheets API
    scope = ["https://docs.google.com/document/d/e/2PACX-1vSHesOf9hv2sPOntssYrEdubmMQm8lwjfwv6NPjjmIRYs_FOYXtqrYgjh85jBUebK9swPXh_a5TJ5Kl/pub"]
    creds = ServiceAccountCredentials.from_json_keyfile_name('credentials.json', scope)
    client = gspread.authorize(creds)

    # Extract the sheet ID from the URL
    sheet_id = url.split('/d/')[1].split('/')[0]

    # Open the Google Sheet by ID
    sheet = client.open_by_key(sheet_id).sheet1

    # Get all records from the sheet
    data = sheet.get_all_records()
    return data
```

```python
def create_and_print_grid(data):
    # Parse the data to find the grid dimensions
    max_x = max(item['x'] for item in data)
    max_y = max(item['y'] for item in data)

    # Initialize the grid with spaces
    grid = [[' ' for _ in range(max_x + 1)] for _ in range(max_y + 1)]

    # Fill the grid with characters from the data
    for item in data:
        x = item['x']
        y = item['y']
        char = item['character']
        grid[y][x] = char

    # Print the grid
    for row in grid:
        print(''.join(row))
```

```python
def main(url):
    data = get_google_sheet_data(url)
    create_and_print_grid(data)
```
