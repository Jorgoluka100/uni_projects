```python
from pyspark.ml.feature import StringIndexer
columns_to_convert = ["period", "bx_gse", "by_gse", "bz_gse", "theta_gse", "phi_gse", "bx_gsm", "source"]
for column in columns_to_convert:
  indexer = StringIndexer(inputCol=column, outputCol=column + "_converted")
  df = indexer.fit(df).transform(df)

df.show(20)
```

```python
df.printSchema()
```

```python
from pyspark.ml.feature import VectorAssembler
features_cols = ["by_gsm", "bz_gsm", "theta_gsm", "phi_gsm", "bt", "density", "speed", "temperature", "bx_gse_converted", "by_gse_converted",
                 "bz_gse_converted", "theta_gse_converted", "phi_gse_converted", "bx_gsm_converted"]

# Create a vectorassembler with all input features
assembler = VectorAssembler(inputCols = features_cols, outputCol = 'features')
data = assembler.transform(df)

data = data.select('features')

data.show(5, truncate = False)
```

```python
from pyspark.ml.feature import RobustScaler, StandardScaler

data.withColumn("scaledFeatures", lit(0)).show()

# # StandardScaler: (X - avg) / Std
# #scaler = StandardScaler(inputCol = 'features', outputCol = 'scaledFeatures')

# #MinMax: (X - min) / (max - min)
# #scaler = MinMaxScaler(inputCol = 'features', outputCol = 'scaledFeatures')

# robustScaler: robust against outliers: (X - med) / IQR; [IQR = Q3 - Q1]
scaler = RobustScaler(inputCol = 'features', outputCol = 'scaledFeatures')

scaler_model = scaler.fit(data)
data = scaler_model.transform(data)

data = data.select("scaledFeatures")
data.show(3, truncate = False)
```

```python
from pyspark.ml.evaluation import ClusteringEvaluator
from pyspark.ml.clustering import KMeans, KMeansModel
silhouette_score=[]
evaluator = ClusteringEvaluator(predictionCol='prediction', featuresCol='features', \
                                metricName='silhouette', distanceMeasure='squaredEuclidean')
for i in range(2,10):
    kmeans=KMeans(featuresCol='features', k=i)
    model=kmeans.fit(data)
    prediction=model.transform(data)
    score=evaluator.evaluate(prediction)
    silhouette_score.append(score)
    print('Silhouette Score for k =',i,'is',score)
```
