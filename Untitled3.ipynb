```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv("heatstroke.csv")
data
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
string_columns = ["experiment", "Task", "Thermal sensation", "Personal Thermal Assessment"]
for column in string_columns:
  data[column] = le.fit_transform(data[column])
```

```python
data1 = data.iloc[:, [1,6 ] ]
data2 = data.iloc[:, [2,6 ] ]
data3 = data.iloc[:, [3,7 ] ]
data4 = data.iloc[:, [2,8 ] ]
data5 = data.iloc[:, [5,8 ] ]
```

```python
from matplotlib import pyplot as plt
data3['Thermal sensation'].plot(kind='hist', bins=20, title='Thermal sensation')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
data3['hrv_entropy'].plot(kind='hist', bins=20, title='hrv_entropy')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
data_result1 = KMeans()
data_result2 = KMeans()
data_result3 = KMeans()
data_result4 = KMeans()
data_result5 = KMeans()
```

```python
KElbowVisualizer(data_result1, k = (1,10)).fit(data1)
```

```python
data2 = KMeans(n_clusters=3, init = 'k-means++',random_state=10).fit(data1)
```

```python
KElbowVisualizer(data_result2, k = (1,10)).fit(data2)
```

```python
code_mapping = {
    'neutral': 1,
    'slightly warm': 2,
    'warm': 3,
    'hot': 4}

column_to_replace = 'Thermal sensation'
data[column_to_replace] = data[column_to_replace].map(code_mapping)
```
