```python
import pandas as pd
#import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
# read csv file, and assign it to the variable
data = pd.read_csv("heatstroke.csv")
data
```

```python
# converting text from column values to numbers
le = LabelEncoder()
string_columns = ["experiment", "Task", "Thermal sensation", "Personal Thermal Assessment"]
# for loop that iterates through all string_columns one by one
for column in string_columns:
  data[column] = le.fit_transform(data[column])

data
```

```python
# specify 2 columns that you want to compare from dataset and assign it to the variable
hrv_mean_Temperature_data = data.iloc[:, [1,6 ] ]
hrv_std_Temperature_data  = data.iloc[:, [2,6 ] ]
hrv_entropy_Thermal_sensation_data  = data.iloc[:, [3,7 ] ]
hrv_std_Personal_Thermal_Assessment_data = data.iloc[:, [2,8 ] ]
Task_Personal_Thermal_Assessment_data = data.iloc[:, [5,8 ] ]
```

```python
# Compare hrv_mean with temperature data using scatterplot
sns.scatterplot(hrv_mean_Temperature_data, x = "hrv_mean", y = "Temperature")
plt.title("hrv_mean and Temperature comparison")
plt.show()
```

```python
# Compare hrv_mean and temperature comparison data using line graph
hrv_mean_Temperature_data.plot(kind='line', figsize=(8, 4), title='hrv_mean and Temperature comparison')
plt.xlabel("ID")
plt.ylabel("Value")
```

```python
# Compare hrv_std with temperature data using scatterplot
sns.scatterplot(hrv_std_Temperature_data, x = "hrv_std", y = "Temperature")
plt.title("hrv_std and Personal Temperature comparison")
plt.show()
```

```python
# Compare hrv_std and Temperature comparison data using line graph
hrv_std_Temperature_data.plot(kind='line', figsize=(8, 4), title='hrv_std and Temperature comparison')
plt.xlabel("ID")
plt.ylabel("Value")
```

```python
# Compare hrv_entropy with thermal_sensation data using scatterplot
sns.scatterplot(hrv_entropy_Thermal_sensation_data, x = "hrv_entropy", y = "Thermal sensation")
plt.title("hrv_entropy and Personal Thermal sensation comparison")
plt.show()
```

```python
# Compare hrv_entropy and Thermal sensation data using line graph
hrv_entropy_Thermal_sensation_data.plot(kind='line', figsize=(8, 4), title='hrv_entropy and Thermal sensation')
plt.xlabel("ID")
plt.ylabel("Value")
```
