```python
# Linking with Spark
import pyspark
from pyspark.sql import SparkSession

spark = SparkSession.builder \
                    .appName("Tutorial_DF") \
                    .master("local[*]") \
                    .config("spark.executor.memory", "4g") \
                    .config("spark.driver.memory", "2g") \
                    .config("spark.executor.cores", "2") \
                    .config("spark.sql.inMemoryColumnarStorage.compressed", "true") \
                    .getOrCreate()

spark
```

```python
df = spark.read.format('csv').load('/content/Tutorial2_PySpark_DF_CN7030_(uel).ipynb', inferSchema = True, header = True)

# show table
df.show(truncate = True)

# show schema
df.printSchema()

# some info
print(df.count())
print(len(df.columns))
```
