```python
def is_right_operator(stringToCheck):
  return (stringToCheck == "+" or stringToCheck == "-" or stringToCheck == "/" or stringToCheck == "*")

def perform_caluclation(number1, number2, operator):
  if operator == "+":
    return number1 + number2
  elif operator == "-":
    return number1 - number2
  elif operator == "/":
    return number1 / number2
  elif operator == "*":
    return number1 * number2
  else:
    print("We shouldn't get here")

def is_right_string(stringToCheck):
  return (stringToCheck.lower() == "y" or stringToCheck.lower() == "n")

player_has_quitted = False
while (player_has_quitted is not True):
    number1 = input("Type first number\n")
    number2 = input("Type second number\n")
    operator = input("What operator do you want to use? Choose: +, -, *, /.\n")
    while (is_right_operator(operator) is False):
      operator = input("Choose correct operator. Choose: +, -, *, /.\n")

    print(f"{number1} {operator} {number2} = {perform_caluclation(int(number1), int(number2), operator)}")
    go_again = input("Do you want to do another calulcation? Y or N\n")
    while (is_right_string(go_again) is False):
          go_again = input("Type Y or N")

    if (go_again.lower() == "n"):
      break


```
